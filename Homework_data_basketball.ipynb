```python
def fplotagem(qualdado:np.array=Games, Jogadores:list=Players):
    colorline={'KobeBryant': 'Orange', 'JoeJohnson': 'Navy', 'LeBronJames': 'Brown', 'CarmeloAnthony': 'Indigo', 'DwightHoward': 'Gray', 'ChrisBosh': 'Green', 'ChrisPaul': 'Red', 'KevinDurant': 'Yellow', 'DerrickRose': 'Blue', 'DwayneWade': 'Black'}
    mk1={'KobeBryant': '^', 'JoeJohnson': 's', 'LeBronJames': 's', 'CarmeloAnthony': 'D', 'DwightHoward': 's', 'ChrisBosh': 'D', 'ChrisPaul': '^', 'KevinDurant': 'D', 'DerrickRose': 's', 'DwayneWade': 'o'}
    for i in Jogadores:
        plt.plot(qualdado[dic_players[i]],c=colorline[i],ls=':', marker=mk1[i],label=i)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.title('Grafico', fontdict=None, loc='center', pad=None)
    plt.show()
```
