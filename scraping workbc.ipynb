```python

#nao ovu usar requests pois o site usa javascript para mostrar a tabela de vagas
#entao vou usar o selenium para abrir a pagina, fazer a request e devolver o HTML
#import requests


import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
```

```python
url_s ='https://www.workbc.ca/Jobs-Careers/Find-Jobs/Jobs.aspx#/job-search;city=Victoria;'
```

```python
#nao vamos usar requests
# pagina = requests.get(url_s)
# html = pagina.text

browser = webdriver.Chrome('C:\Program Files (x86)\chromedriver.exe')
browser.get(url_s)
html = browser.page_source
```

```python
time.sleep(4)
df = pd.DataFrame(columns=['Cargo','Empresa','Local', 'Fonte','salario','Tipo','Postada em'])
soup = BeautifulSoup(html, 'lxml')
blocojobs = soup.find('div', class_='job-listings ng-star-inserted')
    
jobs = blocojobs.find_all('div',class_='job-item print-no-break')
    
print(browser.current_url)
for j in jobs:
    newline=[j.find('div',class_='title').text,j.find('div',class_='company ng-star-inserted').text,\
                 j.find('div',class_='location ng-star-inserted').text[10:],\
                 j.find('div',class_='ng-star-inserted').text,\
                 j.find('div',class_='salary').text[7:],\
                 j.find('div',class_='position-type ng-star-inserted').text[7:],\
                 j.find('div',class_='ng-star-inserted').text[8:19]]
    tam=len(df)
    df.loc[tam]=newline
```

```python
quantidadepaginas = soup.find('div', class_='page-of ng-star-inserted').text
quantidadepaginas=int(quantidadepaginas.split()[3])
quantidadepaginas
```

```python
for nr_pagina in range(0,quantidadepaginas-1):
    
    browser.find_element(By.XPATH, '//*[@id="form"]/div[5]/main/div/app-root/div/app-job-search/app-results/div/div/div[5]/app-pagination/div[2]/div[7]/div[1]/button').click()
    time.sleep(2)
    html = browser.page_source
    
    soup = BeautifulSoup(html, 'lxml')
    
    blocojobs = soup.find('div', class_='job-listings ng-star-inserted')
    
    jobs = blocojobs.find_all('div',class_='job-item print-no-break')
    
    print(browser.current_url)
    for j in jobs:
        newline=[j.find('div',class_='title').text,j.find('div',class_='company ng-star-inserted').text,\
                 j.find('div',class_='location ng-star-inserted').text[10:],\
                 j.find('div',class_='ng-star-inserted').text,\
                 j.find('div',class_='salary').text[7:],\
                 j.find('div',class_='position-type ng-star-inserted').text[7:],\
                 j.find('div',class_='ng-star-inserted').text[8:19]]
        tam=len(df)
        df.loc[tam]=newline
df.to_csv('empregosdoworkbc.csv')       
        
```
