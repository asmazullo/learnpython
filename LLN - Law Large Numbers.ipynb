Law of Large Numbers
___
"the average of the results obtained from a large number of trials should be close to the expected value and tends to become closer to the expected value as more trials are performed"


I will check the LLN theory using a funcion called randn(). This funtion chose a number based on a normal distribution.

Using a loop function, I will count the numbers between -1 and 1 and divide it by the total random numbers. 

The expected result is 68,2%

I will increase the total number from 10 until 10000000 

```python
#importing RANDN from numpy
#RANDN returns a randon number based in a normal distribution
```

```python
#randon normal numbers using 10 numbers
total = 10
#counter
counter = 0
for i in range(0, total):
    number = randn()
    if 1>= number >= -1:
        counter = counter+1
print ("average = ",counter/total)
    
```
